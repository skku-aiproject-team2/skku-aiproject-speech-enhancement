```python
# choose 10% of files in train/noisy and train/clean and move them to dataset_mini/train/noisy and dataset_mini/train/clean
# file name should be the same in noisy and clean

import os
import random
import shutil


src_noisy = '../dataset/train/noisy'
src_clean = '../dataset/train/clean'
dst_noisy = '../dataset_mini/train/noisy'
dst_clean = '../dataset_mini/train/clean'

os.makedirs(dst_noisy, exist_ok=True)
os.makedirs(dst_clean, exist_ok=True)

noisy_files = os.listdir(src_noisy)
clean_files = os.listdir(src_clean)

random.shuffle(noisy_files)

for i in range(int(len(noisy_files) * 0.1)):
    file = noisy_files[i]
    shutil.copy(os.path.join(src_noisy, file), os.path.join(dst_noisy, file))
    shutil.copy(os.path.join(src_clean, file), os.path.join(dst_clean, file))
    
    # print from and to
    print(os.path.join(src_noisy, file), os.path.join(dst_noisy, file))
    print(os.path.join(src_clean, file), os.path.join(dst_clean, file))
    print('')
```
